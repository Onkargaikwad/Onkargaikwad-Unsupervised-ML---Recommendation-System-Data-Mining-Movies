```
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```
# Import Dataset
movies=pd.read_csv('my_movies.csv')
movies
```

```
# Sort by User IDs
movies.sort_values(['V1'])
```

```
len(movies['V1'].unique())
```

```
# number of unique movies in the dataset
len(movies['V5'].unique())
```

```
# converting long data into wide data using pivot
movies2=movies.pivot_table(index='V1',columns='V2').reset_index(drop=True)
movies2
```

```

# Impute those NaNs with 0 values
movies2.fillna(0,inplace=True)
movies2
```

```

# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(movies2.values,metric='cosine')
user_sim
```

```

# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```
# Slicing first 5 rows and first 5 columns
user_sim2.iloc[0:5,0:5]
```

```
# Most Similar Users
user_sim2.idxmax(axis=1)[0:5]
```
